```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

def focal_loss(y_true, y_pred, gamma=3.0, alpha=0.90):
    y_true = tf.cast(y_true, tf.float32)
    epsilon = 1e-7
    y_pred = tf.clip_by_value(y_pred, epsilon, 1. - epsilon)

    p_t = tf.where(tf.equal(y_true, 1), y_pred, 1 - y_pred)
    alpha_t = tf.where(tf.equal(y_true, 1), alpha, 1 - alpha)
    focal_weight = alpha_t * tf.pow(1 - p_t, gamma)

    ce = -y_true * tf.math.log(y_pred) - (1 - y_true) * tf.math.log(1 - y_pred)
    return tf.reduce_mean(focal_weight * ce)

def soft_fbeta_loss(y_true, y_pred, beta=2.5, eps=1e-7):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    tp = tf.reduce_sum(y_true * y_pred)
    fp = tf.reduce_sum((1.0 - y_true) * y_pred)
    fn = tf.reduce_sum(y_true * (1.0 - y_pred))

    precision = (tp + eps) / (tp + fp + eps)
    recall = (tp + eps) / (tp + fn + eps)

    b2 = beta ** 2
    fbeta = (1.0 + b2) * precision * recall / (b2 * precision + recall + eps)
    return 1.0 - fbeta

def recall_focused_loss(y_true, y_pred, fn_weight=15.0, eps=1e-7):
    y_true = tf.cast(y_true, tf.float32)
    y_pred = tf.cast(y_pred, tf.float32)

    bce = -fn_weight * y_true * tf.math.log(y_pred + eps) - (1 - y_true) * tf.math.log(1 - y_pred + eps)
    return tf.reduce_mean(bce)


class OptimizedSafetyPINN(keras.Model):
    def __init__(self, phys_idx, input_dim, focal_alpha=0.90, focal_gamma=3.0,
                 f2_weight=2.5, recall_weight=1.0, dropout_rate=0.25,
                 clip_norm=1.0, beta=2.5, phys_warmup_epochs=10,
                 max_phys_weight=0.08, phys_data_weight=0.05):
        super().__init__()
        self.phys_idx = phys_idx
        self.input_dim = input_dim

        # Safety-focused hyperparameters
        self.focal_alpha = focal_alpha
        self.focal_gamma = focal_gamma
        self.f2_weight = f2_weight
        self.recall_weight = recall_weight
        self.clip_norm = clip_norm
        self.beta = beta

        # Physics integration parameters
        self.phys_warmup_epochs = phys_warmup_epochs
        self.max_phys_weight = max_phys_weight
        self.phys_data_weight = phys_data_weight
        self.current_epoch = tf.Variable(0, dtype=tf.int32, trainable=False)

        # Attention layer
        self.attention_dense = layers.Dense(input_dim, activation='tanh',
                                           kernel_regularizer=keras.regularizers.l2(1e-4))
        self.attention_weights = layers.Dense(input_dim, activation='softmax')

        # Deep network with residual connections
        self.proj1 = layers.Dense(256)
        self.dense1 = layers.Dense(256, kernel_regularizer=keras.regularizers.l2(1e-4))
        self.bn1 = layers.BatchNormalization()
        self.drop1 = layers.Dropout(dropout_rate)

        self.dense2 = layers.Dense(256, kernel_regularizer=keras.regularizers.l2(1e-4))
        self.bn2 = layers.BatchNormalization()
        self.drop2 = layers.Dropout(dropout_rate)

        self.proj2 = layers.Dense(128)
        self.dense3 = layers.Dense(128, kernel_regularizer=keras.regularizers.l2(1e-4))
        self.bn3 = layers.BatchNormalization()
        self.drop3 = layers.Dropout(dropout_rate)

        self.dense4 = layers.Dense(128, kernel_regularizer=keras.regularizers.l2(1e-4))
        self.bn4 = layers.BatchNormalization()
        self.drop4 = layers.Dropout(dropout_rate)

        self.proj3 = layers.Dense(64)
        self.dense5 = layers.Dense(64, kernel_regularizer=keras.regularizers.l2(1e-4))
        self.bn5 = layers.BatchNormalization()
        self.drop5 = layers.Dropout(dropout_rate)

        # Avalanche prediction head
        self.aval_dense1 = layers.Dense(64, activation='relu',
                                        kernel_regularizer=keras.regularizers.l2(1e-4))
        self.aval_bn1 = layers.BatchNormalization()
        self.aval_dense2 = layers.Dense(32, activation='relu',
                                        kernel_regularizer=keras.regularizers.l2(1e-4))
        self.aval_dense3 = layers.Dense(16, activation='relu')
        self.aval_head = layers.Dense(1, activation='sigmoid', name='avalanche')

        # Physics head
        self.phys_dense1 = layers.Dense(32, activation='relu')
        self.phys_dense2 = layers.Dense(16, activation='relu')
        self.phys_head = layers.Dense(1, activation='linear', name='temp_change')

        # Learnable physics coefficient
        self.alpha = tf.Variable(0.1, dtype=tf.float32, trainable=True, name='alpha')
        self.loss_fn_mse = keras.losses.MeanSquaredError()

        # Metrics
        self.acc_metric = keras.metrics.BinaryAccuracy(name="accuracy")
        self.prec_metric = keras.metrics.Precision(name="precision")
        self.rec_metric = keras.metrics.Recall(name="recall")
        self.phys_loss_tracker = keras.metrics.Mean(name="phys_eqn_loss")

    @property
    def metrics(self):
        return [self.acc_metric, self.prec_metric, self.rec_metric, self.phys_loss_tracker]

    def get_physics_weight(self):
        epoch = tf.cast(self.current_epoch, tf.float32)
        warmup = tf.cast(self.phys_warmup_epochs, tf.float32)
        weight = tf.minimum(epoch / warmup, 1.0) * self.max_phys_weight
        return weight

    def call(self, inputs, training=False):
        # Attention mechanism
        att = self.attention_dense(inputs)
        att_weights = self.attention_weights(att)
        x = inputs * att_weights

        # Block 1
        x = self.proj1(x)
        x1 = self.dense1(x)
        x1 = self.bn1(x1, training=training)
        x1 = tf.nn.leaky_relu(x1, alpha=0.1)
        x1 = self.drop1(x1, training=training)

        # Block 2 + Residual
        x2 = self.dense2(x1)
        x2 = self.bn2(x2, training=training)
        x2 = tf.nn.leaky_relu(x2, alpha=0.1)
        x2 = x2 + x1
        x2 = self.drop2(x2, training=training)

        # Block 3
        x3 = self.proj2(x2)
        x3 = self.dense3(x3)
        x3 = self.bn3(x3, training=training)
        x3 = tf.nn.leaky_relu(x3, alpha=0.1)
        x3 = self.drop3(x3, training=training)

        # Block 4 + Residual
        x4 = self.dense4(x3)
        x4 = self.bn4(x4, training=training)
        x4 = tf.nn.leaky_relu(x4, alpha=0.1)
        x4 = x4 + x3
        x4 = self.drop4(x4, training=training)

        # Block 5
        x5 = self.proj3(x4)
        x5 = self.dense5(x5)
        x5 = self.bn5(x5, training=training)
        feat = tf.nn.leaky_relu(x5, alpha=0.1)
        feat = self.drop5(feat, training=training)

        # Avalanche head
        aval_x = self.aval_dense1(feat)
        aval_x = self.aval_bn1(aval_x, training=training)
        aval_x = self.aval_dense2(aval_x)
        aval_x = self.aval_dense3(aval_x)
        aval_out = self.aval_head(aval_x)

        # Physics head
        phys_x = self.phys_dense1(feat)
        phys_x = self.phys_dense2(phys_x)
        phys_out = self.phys_head(phys_x)

        return aval_out, phys_out

    def train_step(self, data):
        x, y = data
        y_aval_true = tf.reshape(y[0], (-1, 1))
        y_phys_true = tf.reshape(y[1], (-1, 1))

        with tf.GradientTape() as tape:
            y_aval_pred, y_phys_pred = self(x, training=True)

            # Avalanche losses
            loss_focal = focal_loss(y_aval_true, y_aval_pred,
                                   gamma=self.focal_gamma, alpha=self.focal_alpha)
            loss_f2 = soft_fbeta_loss(y_aval_true, y_aval_pred, beta=self.beta)
            loss_recall = recall_focused_loss(y_aval_true, y_aval_pred, fn_weight=15.0)
            loss_aval = loss_focal + (self.f2_weight * loss_f2) + (self.recall_weight * loss_recall)

            # Physics losses
            loss_phys_data = self.loss_fn_mse(y_phys_true, y_phys_pred)

            def get_tensor(idx):
                if idx is None: return tf.zeros_like(y_phys_pred)
                t = tf.gather(x, idx, axis=1)
                return tf.reshape(t, (-1, 1))

            iswr = get_tensor(self.phys_idx['ISWR'])
            ilwr = get_tensor(self.phys_idx['ILWR'])
            olwr = get_tensor(self.phys_idx['OLWR'])
            qs = get_tensor(self.phys_idx['Qs'])
            ql = get_tensor(self.phys_idx['Ql'])

            # Energy balance: Q_net = ISWR + ILWR - OLWR + Qs + Ql
            net_flux = iswr + ilwr - olwr + qs + ql
            physics_residual = y_phys_pred - (self.alpha * net_flux)
            loss_phys_eqn = tf.reduce_mean(tf.square(physics_residual))

            # Adaptive physics weighting
            phys_weight = self.get_physics_weight()
            loss_phys_scaled = phys_weight * loss_phys_eqn
            loss_phys_data_scaled = self.phys_data_weight * loss_phys_data

            total_loss = loss_aval + loss_phys_data_scaled + loss_phys_scaled

        # Backpropagation
        grads = tape.gradient(total_loss, self.trainable_weights)
        if self.clip_norm is not None:
            grads, _ = tf.clip_by_global_norm(grads, self.clip_norm)
        self.optimizer.apply_gradients(zip(grads, self.trainable_weights))

        # Update metrics
        self.acc_metric.update_state(y_aval_true, y_aval_pred)
        self.prec_metric.update_state(y_aval_true, y_aval_pred)
        self.rec_metric.update_state(y_aval_true, y_aval_pred)
        self.phys_loss_tracker.update_state(loss_phys_eqn)

        p = self.prec_metric.result()
        r = self.rec_metric.result()
        f1 = 2.0 * (p * r) / (p + r + 1e-7)
        f2 = 5.0 * (p * r) / (4.0 * p + r + 1e-7)
        current_phys_weight = self.get_physics_weight()

        return {
            "loss": total_loss, "focal": loss_focal, "f2_loss": loss_f2,
            "accuracy": self.acc_metric.result(), "precision": p,
            "recall": r, "f1": f1, "f2": f2, "phys_eqn_loss": loss_phys_eqn,
            "phys_weight": current_phys_weight, "alpha": self.alpha
        }

    def test_step(self, data):
        x, y = data
        y_aval_true = tf.reshape(y[0], (-1, 1))
        y_phys_true = tf.reshape(y[1], (-1, 1))

        y_aval_pred, y_phys_pred = self(x, training=False)

        # Avalanche losses
        loss_focal = focal_loss(y_aval_true, y_aval_pred,
                               gamma=self.focal_gamma, alpha=self.focal_alpha)
        loss_f2 = soft_fbeta_loss(y_aval_true, y_aval_pred, beta=self.beta)
        loss_recall = recall_focused_loss(y_aval_true, y_aval_pred, fn_weight=15.0)
        loss_aval = loss_focal + self.f2_weight * loss_f2 + self.recall_weight * loss_recall

        # Physics losses
        loss_phys_data = self.loss_fn_mse(y_phys_true, y_phys_pred)

        def get_tensor(idx):
            if idx is None: return tf.zeros_like(y_phys_pred)
            t = tf.gather(x, idx, axis=1)
            return tf.reshape(t, (-1, 1))

        iswr = get_tensor(self.phys_idx['ISWR'])
        ilwr = get_tensor(self.phys_idx['ILWR'])
        olwr = get_tensor(self.phys_idx['OLWR'])
        qs = get_tensor(self.phys_idx['Qs'])
        ql = get_tensor(self.phys_idx['Ql'])

        # Energy balance
        net_flux = iswr + ilwr - olwr + qs + ql
        physics_residual = y_phys_pred - (self.alpha * net_flux)
        loss_phys_eqn = tf.reduce_mean(tf.square(physics_residual))

        phys_weight = self.get_physics_weight()
        total_loss = loss_aval + self.phys_data_weight * loss_phys_data + phys_weight * loss_phys_eqn

        self.acc_metric.update_state(y_aval_true, y_aval_pred)
        self.prec_metric.update_state(y_aval_true, y_aval_pred)
        self.rec_metric.update_state(y_aval_true, y_aval_pred)
        self.phys_loss_tracker.update_state(loss_phys_eqn)

        p = self.prec_metric.result()
        r = self.rec_metric.result()
        f1 = 2.0 * (p * r) / (p + r + 1e-7)
        f2 = 5.0 * (p * r) / (4.0 * p + r + 1e-7)

        return {
            "loss": total_loss, "focal": loss_focal, "f2_loss": loss_f2,
            "accuracy": self.acc_metric.result(), "precision": p,
            "recall": r, "f1": f1, "f2": f2, "phys_eqn_loss": loss_phys_eqn
        }
```

```python
# =====================================================================
# FLEXIBLE INFERENCE WITH KNN IMPUTATION
# Accepts any subset of features and fills in the rest
# =====================================================================

def prepare_partial_input(partial_data, feature_names=FEATURE_NAMES, 
                          scaler=scaler, knn_imputer=knn_imputer):
    
    if isinstance(partial_data, dict):
        partial_data = pd.DataFrame([partial_data])
    
    complete_df = pd.DataFrame(columns=feature_names, index=range(len(partial_data)))
    
    for col in partial_data.columns:
        if col in feature_names:
            complete_df[col] = partial_data[col].values
    
    # Convert to float (NaN stays as NaN)
    complete_df = complete_df.astype(float)
    
    provided_features = [col for col in partial_data.columns if col in feature_names]
    missing_features = [col for col in feature_names if col not in partial_data.columns]
    
    print(f"📊 Input Analysis:")
    print(f"   ✅ Provided: {len(provided_features)}/{len(feature_names)} features")
    print(f"   🔄 Will impute: {len(missing_features)} features using KNN")
    
    if len(missing_features) > 0 and len(missing_features) <= 10:
        print(f"   Missing: {missing_features}")
    elif len(missing_features) > 10:
        print(f"   Missing (first 10): {missing_features[:10]}...")
    
    for col in provided_features:
        idx = feature_names.index(col)
        mean, scale = scaler.mean_[idx], scaler.scale_[idx]
        complete_df[col] = (complete_df[col] - mean) / scale
    
    # Use KNN to impute missing values (works on scaled data)
    X_imputed = knn_imputer.transform(complete_df.values)
    
    return X_imputed


def predict_with_partial_data(partial_data, model=model, threshold=None):
    # Prepare the input
    X_complete = prepare_partial_input(partial_data)
    
    # Get predictions
    aval_prob, phys_pred = model.predict(X_complete, verbose=0)
    aval_prob = aval_prob.flatten()[0]
    
    # Use default threshold if not specified
    if threshold is None:
        threshold = best_thresh  # From training optimization
    
    # Classification
    is_avalanche = aval_prob >= threshold
    
    # Risk level
    if aval_prob < 0.3:
        risk_level = "🟢 LOW"
    elif aval_prob < 0.5:
        risk_level = "🟡 MODERATE"
    elif aval_prob < 0.7:
        risk_level = "🟠 CONSIDERABLE"
    elif aval_prob < 0.85:
        risk_level = "🔴 HIGH"
    else:
        risk_level = "⚫ EXTREME"
    
    print(f"\n{'='*50}")
    print(f"🏔️  AVALANCHE PREDICTION RESULT")
    print(f"{'='*50}")
    print(f"   Probability: {aval_prob:.1%}")
    print(f"   Risk Level:  {risk_level}")
    print(f"   Threshold:   {threshold:.2f}")
    print(f"   Prediction:  {'⚠️ AVALANCHE LIKELY' if is_avalanche else '✅ LOW RISK'}")
    print(f"{'='*50}")
    
    return {
        'probability': float(aval_prob),
        'is_avalanche': bool(is_avalanche),
        'risk_level': risk_level,
        'threshold_used': threshold,
        'physics_prediction': float(phys_pred.flatten()[0])
    }


# =====================================================================
# EXAMPLE USAGE
# =====================================================================
print("="*70)
print("📘 FLEXIBLE INPUT SYSTEM READY")
print("="*70)
print("\nExample usage with partial data:")
print("""
# Provide only the features you have:
partial_input = {
    'TA': -5.0,           # Air temperature (°C)
    'max_height': 150.0,  # Snow height (cm)
    'mean_lwc': 0.02,     # Mean liquid water content
    'S5': 1.5,            # Stability index
}

# Get prediction (missing features filled by KNN)
result = predict_with_partial_data(partial_input)
""")
print("="*70)
```

```python
# =====================================================================
# USER INPUT: Paste your data in "param value, param2 value2" format
# =====================================================================

def parse_input_string(input_string):
    """
    Parse a comma-separated string of 'parameter value' pairs into a dictionary.
    
    Example input:
    "TA 8.3000, TA_daily -0.1000, ISWR_daily 156.00, max_height 0.2400"
    
    Returns:
    dict: {'TA': 8.3, 'TA_daily': -0.1, 'ISWR_daily': 156.0, 'max_height': 0.24}
    """
    data = {}
    
    # Split by comma
    pairs = input_string.strip().split(',')
    
    for pair in pairs:
        pair = pair.strip()
        if not pair:
            continue
            
        # Split by whitespace - parameter name is first, value is second
        parts = pair.split()
        if len(parts) >= 2:
            param_name = parts[0].strip()
            try:
                value = float(parts[1].strip())
                data[param_name] = value
            except ValueError:
                print(f"⚠️ Could not parse value for '{param_name}': {parts[1]}")
        else:
            print(f"⚠️ Invalid format: '{pair}' - expected 'parameter value'")
    
    return data


# =============================================================================
# 📝 PASTE YOUR DATA BELOW (between the triple quotes)
# =============================================================================
user_input = """
TA -13.0000, TA_daily -13.0000, ISWR_daily 0.4608, ISWR_dir_daily 0.2765, ISWR_diff_daily 0.1843, ILWR_daily 194.7779, max_height 1.0305, SWE_daily 18.2000, MS_Rain_daily 3.5700, Qs 1608.0000, Ql 48.2400, water -65.0000, max_lwc -9.4614, S5 3.0000
"""

# Parse the input
parsed_data = parse_input_string(user_input)

print("="*70)
print("📥 PARSED INPUT DATA")
print("="*70)
print(f"Found {len(parsed_data)} parameters:\n")
for param, value in parsed_data.items():
    print(f"  {param}: {value}")

# Run prediction if data was parsed
if parsed_data:
    print("\n" + "="*70)
    print("🔮 RUNNING PREDICTION...")
    print("="*70)
    result = predict_with_partial_data(parsed_data)
else:
    print("\n❌ No valid data found. Please check your input format.")
```

```python
from sklearn.metrics import roc_curve, auc, roc_auc_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# New test data (data3 + data4)
X_new_test = test_data[existing_features]
X_new_test = pd.DataFrame(simple_imputer.transform(X_new_test), columns=X_new_test.columns)
X_new_test = pd.DataFrame(scaler.transform(X_new_test), columns=X_new_test.columns)
y_new_test = test_data['avalancheDay1']

y_new_pred_prob, _ = model.predict(X_new_test)
y_new_prob_flat = y_new_pred_prob.flatten()

# Calculate ROC curve
fpr, tpr, thresholds = roc_curve(y_new_test, y_new_prob_flat)
roc_auc = auc(fpr, tpr)

# Find optimal threshold using Youden's J statistic (balances TPR and FPR)
j_scores = tpr - fpr
optimal_idx = np.argmax(j_scores)
optimal_thresh = thresholds[optimal_idx]

# Find most accurate threshold
best_accuracy = 0
accuracy_thresh = 0.5
for thresh in np.arange(0.05, 0.95, 0.01):
    y_pred_temp = (y_new_prob_flat >= thresh).astype(int)
    acc = (y_pred_temp == y_new_test).mean()
    if acc > best_accuracy:
        best_accuracy = acc
        accuracy_thresh = thresh

# ============================================
# Plot ROC curve
# ============================================
plt.figure(figsize=(10, 8))

plt.plot(fpr, tpr, color='green', lw=2,
         label=f'Test Data (AUC = {roc_auc:.4f})')

# Random classifier line
plt.plot([0, 1], [0, 1], color='red', lw=2, linestyle='--', label='Random Classifier')

# Mark optimal threshold points
plt.scatter(fpr[optimal_idx], tpr[optimal_idx],
            marker='o', color='darkgreen', s=150, zorder=5,
            label=f'Youden\'s J Threshold = {optimal_thresh:.2f}')

# Find index for accuracy threshold on ROC curve
acc_thresh_idx = np.argmin(np.abs(thresholds - accuracy_thresh))
plt.scatter(fpr[acc_thresh_idx], tpr[acc_thresh_idx],
            marker='s', color='blue', s=150, zorder=5,
            label=f'Max Accuracy Threshold = {accuracy_thresh:.2f}')

plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (1 - Specificity)', fontsize=12)
plt.ylabel('True Positive Rate (Recall/Sensitivity)', fontsize=12)
plt.title('ROC Curve - Avalanche Prediction Model', fontsize=14, fontweight='bold')
plt.legend(loc='lower right', fontsize=10)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('roc_curve.png', dpi=150, bbox_inches='tight', transparent=True)
plt.show()

# ============================================
# Confusion Matrices Comparison (Poster-Sized)
# ============================================
fig, axes = plt.subplots(1, 2, figsize=(16, 9))
fig.patch.set_alpha(0)
axes[0].patch.set_alpha(0)
axes[1].patch.set_alpha(0)

# --- Confusion Matrix 1: Youden's J Optimal Threshold ---
y_pred_youden = (y_new_prob_flat >= optimal_thresh).astype(int)
cm_youden = confusion_matrix(y_new_test, y_pred_youden)
tn_y, fp_y, fn_y, tp_y = cm_youden.ravel()

sns.heatmap(cm_youden, annot=True, fmt='d', cmap='Blues', ax=axes[0],
            xticklabels=['Predicted Safe', 'Predicted Avalanche'],
            yticklabels=['Actual Safe', 'Actual Avalanche'],
            annot_kws={'size': 22, 'weight': 'bold'},
            cbar_kws={'label': 'Count', 'shrink': 0.8})
axes[0].tick_params(labelsize=14)
axes[0].set_xticklabels(axes[0].get_xticklabels(), fontsize=14, weight='bold')
axes[0].set_yticklabels(axes[0].get_yticklabels(), fontsize=14, weight='bold')
axes[0].set_title(f"Youden's J Threshold = {optimal_thresh:.2f}\n(Balanced TPR/FPR)", fontsize=16, fontweight='bold', pad=20)

recall_y = tp_y / (tp_y + fn_y) if (tp_y + fn_y) > 0 else 0
precision_y = tp_y / (tp_y + fp_y) if (tp_y + fp_y) > 0 else 0
accuracy_y = (tp_y + tn_y) / (tp_y + tn_y + fp_y + fn_y)

# --- Confusion Matrix 2: Maximum Accuracy Threshold ---
y_pred_acc = (y_new_prob_flat >= accuracy_thresh).astype(int)
cm_acc = confusion_matrix(y_new_test, y_pred_acc)
tn_a, fp_a, fn_a, tp_a = cm_acc.ravel()

sns.heatmap(cm_acc, annot=True, fmt='d', cmap='Greens', ax=axes[1],
            xticklabels=['Predicted Safe', 'Predicted Avalanche'],
            yticklabels=['Actual Safe', 'Actual Avalanche'],
            annot_kws={'size': 22, 'weight': 'bold'},
            cbar_kws={'label': 'Count', 'shrink': 0.8})
axes[1].tick_params(labelsize=14)
axes[1].set_xticklabels(axes[1].get_xticklabels(), fontsize=14, weight='bold')
axes[1].set_yticklabels(axes[1].get_yticklabels(), fontsize=14, weight='bold')
axes[1].set_title(f'Max Accuracy Threshold = {accuracy_thresh:.2f}\n(Highest Overall Accuracy)', fontsize=16, fontweight='bold', pad=20)

recall_a = tp_a / (tp_a + fn_a) if (tp_a + fn_a) > 0 else 0
precision_a = tp_a / (tp_a + fp_a) if (tp_a + fp_a) > 0 else 0
accuracy_a = (tp_a + tn_a) / (tp_a + tn_a + fp_a + fn_a)

plt.tight_layout()
plt.savefig('confusion_matrices_comparison.png', dpi=150, bbox_inches='tight', transparent=True)

plt.show()

# ============================================
# Print Detailed Results
# ============================================
print("\n" + "="*70)
print("📊 ROC-AUC SCORE: {:.4f}".format(roc_auc))
print("="*70)

print("\n" + "-"*70)
print("🎯 YOUDEN'S J THRESHOLD (Balanced TPR/FPR)")
print("-"*70)
print(f"Threshold: {optimal_thresh:.3f}")
print(f"Confusion Matrix:")
print(cm_youden)
print(f"\n✅ True Negatives:  {tn_y:,}")
print(f"✅ True Positives:  {tp_y:,}")
print(f"⚠️  False Positives: {fp_y:,}")
print(f"❌ False Negatives: {fn_y:,}")
print(f"\nRecall: {recall_y:.1%} | Precision: {precision_y:.1%} | Accuracy: {accuracy_y:.1%}")

print("\n" + "-"*70)
print("🎯 MAXIMUM ACCURACY THRESHOLD")
print("-"*70)
print(f"Threshold: {accuracy_thresh:.3f}")
print(f"Confusion Matrix:")
print(cm_acc)
print(f"\n✅ True Negatives:  {tn_a:,}")
print(f"✅ True Positives:  {tp_a:,}")
print(f"⚠️  False Positives: {fp_a:,}")
print(f"❌ False Negatives: {fn_a:,}")
print(f"\nRecall: {recall_a:.1%} | Precision: {precision_a:.1%} | Accuracy: {accuracy_a:.1%}")
print("="*70)
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_avalanche_pinn_advanced_diagram():
    fig, ax = plt.subplots(figsize=(18, 11))
    fig.patch.set_alpha(0.0)
    ax.set_xlim(0, 12)
    ax.set_ylim(0, 10)
    ax.set_axis_off()

    # Define helper function for boxes
    def add_box(x, y, w, h, text, title="", color="white", edge="black", title_fontsize=14, text_fontsize=14, title_color='black'):
        rect = patches.FancyBboxPatch((x, y), w, h, boxstyle="round,pad=0.1", 
                                      linewidth=2.5, edgecolor=edge, facecolor=color)
        ax.add_patch(rect)
        if title:
            ax.text(x + w/2, y + h - 0.35, title, ha='center', va='center', fontsize=title_fontsize, fontweight='bold', color=title_color)
        ax.text(x + w/2, y + h/2 - (0.15 if title else 0), text, ha='center', va='center', fontsize=text_fontsize, wrap=True)

    # 1. INPUTS - 38 Environmental Features
    input_text = "38 Environmental\nData Features\nWeather & Snow\nProperties"
    add_box(0.3, 4, 1.8, 2.2, input_text, "INPUTS", "#e3f2fd", "#1565c0", title_fontsize=14, text_fontsize=16)

    # 2. NEURAL NETWORK - Multitask Learning + Residual Blocks
    nn_text = ("Multitask Learning +\n3 Residual Blocks\n(256→128→64)\n\nLearns relationships\nbetween snow/weather\npatterns")
    add_box(2.8, 3.5, 2.6, 3.2, nn_text, 
            "NEURAL NETWORK\n(OptimizedSafetyPINN)", "#fff9c4", "#fbc02d", title_fontsize=14, text_fontsize=16)

    # 3. TOTAL LOSS BOX - Connected from NN, BEFORE output
    loss_equation = (
        "$L_{total} = \\frac{1}{n}\\sum(\\lambda_1 L_{focal} + \\lambda_2 L_{fbeta}$\n"
        "$+ \\lambda_3 L_{recall} + \\lambda_4 L_{physics})$\n\n"
        "Weighted Mean Squared Error\nof all loss components"
    )
    add_box(6.2, 4, 2.8, 3.2, loss_equation, 
            "TOTAL LOSS\n(MSE with λ weights)", "#ffebee", "#b71c1c", title_fontsize=14, text_fontsize=16)

    # 4. OUTPUT - Avalanche Probability
    add_box(9.8, 5.2, 1.8, 1.4, "Avalanche\nProbability\n(0 to 1)", "OUTPUT", "#c8e6c9", "#2e7d32", title_fontsize=14, text_fontsize=16)

    # 5. PHYSICS INFORMATION - Going into Total Loss
    physics_text = (
        "PINN Physics-Informed Constraint\n"
        "$\\frac{\\partial T}{\\partial t} = \\alpha \\cdot (ISWR + ILWR - OLWR + Q_s + Q_l)$\n"
        "Enforces physics constraints"
    )
    add_box(6.2, 1.15, 3.2, 2.0, physics_text, 
            "PHYSICS INFORMATION\n(PINN Constraint)", "#f1f8e9", "#33691e", title_fontsize=14, text_fontsize=15)

    # Connectors (Arrows)
    arrow_props_black = dict(arrowstyle='->', lw=2.5, color='black')
    arrow_props_green = dict(arrowstyle='->', lw=2.5, color='#33691e')
    
    # Input -> NN
    ax.annotate('', xy=(2.8, 5.1), xytext=(2.1, 5.1), arrowprops=arrow_props_black)
    
    # NN -> Total Loss
    ax.annotate('', xy=(6.2, 5.6), xytext=(5.4, 5.1), arrowprops=arrow_props_black)
    
    # Total Loss -> Output
    ax.annotate('', xy=(9.8, 5.9), xytext=(9, 5.6), arrowprops=arrow_props_black)
    
    # Physics -> Total Loss
    ax.annotate('', xy=(7.6, 4.0), xytext=(7.8, 3.15), arrowprops=arrow_props_green)
    ax.text(8.0, 3.5, 'Physics Error', fontsize=13, ha='center', fontweight='bold', color='#33691e',
            bbox=dict(boxstyle='round', facecolor='white', alpha=0.9))

    # Main Title
    plt.text(6, 9.4, "PHYSICS-INFORMED NEURAL NETWORK (PINN) FOR AVALANCHE PREDICTION", 
             fontsize=18, fontweight='bold', ha='center', color='#0d1b2a')

    # Subtitle
    plt.text(6, 8.95, "Multitask Learning with Physics Constraints Integration", 
             fontsize=13, style='italic', ha='center', color='#1a237e')

    # Key Information Box (Bottom)
    key_info = (
        "Key Components: Focal Loss (handles class imbalance) | F-Beta Score (recall-weighted) | "
        "Recall Loss (minimizes false negatives) | Physics Loss (energy balance constraint) | "
        "λ weights (learnable loss balancing)"
    )
    ax.text(6, -0.5, key_info, fontsize=11, ha='center', style='italic', color='#5d4037',
            bbox=dict(boxstyle='round', facecolor='lightyellow', alpha=0.7))

    plt.tight_layout()
    plt.savefig('avalanche_pinn_advanced_architecture.png', dpi=300, bbox_inches='tight', transparent=True)
    plt.show()

    print("="*80)
    print("ADVANCED AVALANCHE PINN ARCHITECTURE DIAGRAM")
    print("="*80)
    print("\n📊 Diagram Components:")
    print("  ✓ Inputs: 38 Environmental Data Features")
    print("  ✓ Neural Network: Multitask Learning + 3 Residual Blocks (256→128→64)")
    print("  ✓ Total Loss: MSE with λ-weighted components (Focal + F-Beta + Recall + Physics)")
    print("  ✓ Physics Information: PINN Constraint enforcing energy balance")
    print("  ✓ Output: Avalanche Probability (0 to 1)")
    print("\n📈 Loss Components:")
    print("  • Focal Loss (γ): Addresses class imbalance in avalanche events")
    print("  • F-Beta Loss (β): Weighted recall-based loss")
    print("  • Recall Loss: Minimizes false negatives (missed avalanches)")
    print("  • Physics Loss: Enforces ΔT = α·Q_net energy balance constraint")
    print("  • λ weights: Learnable coefficients for loss balancing")
    print("\n🎯 Total Loss Equation:")
    print("  L_total = (1/n)Σ[λ₁·L_focal + λ₂·L_fbeta + λ₃·L_recall + λ₄·L_physics]")
    print("="*80)

draw_avalanche_pinn_advanced_diagram()
```

```python
def draw_loss_function_breakdown():
    fig, ax = plt.subplots(figsize=(10, 10))
    fig.patch.set_alpha(0.0)
    ax.set_xlim(0, 10)
    ax.set_ylim(0, 10)
    ax.set_axis_off()

    # Main Equation Box
    loss_equation = (
        "$L_{total} = \\frac{1}{n}\\sum(\\lambda_1 L_{focal} + \\lambda_2 L_{fbeta} + \\lambda_3 L_{recall} + \\lambda_4 L_{physics})$"
    )
    
    # Draw huge equation in the center top
    ax.text(5, 8.5, "Total Loss Function", fontsize=20, fontweight='bold', ha='center', color='#0d1b2a')
    
    rect = patches.FancyBboxPatch((0.5, 6.5), 9, 1.5, boxstyle="round,pad=0.2", 
                                  linewidth=3, edgecolor='#b71c1c', facecolor='#ffebee')
    ax.add_patch(rect)
    ax.text(5, 7.25, loss_equation, ha='center', va='center', fontsize=22)

    # Variable Explanations
    y_start = 5.5
    y_step = 0.9
    
    variables = [
        (r"$L_{total}$", "Final weighted loss value to be minimized"),
        (r"$n$", "Number of samples in the batch"),
        (r"$\lambda_{1-4}$", "Learnable weights that balance the importance of each loss component"),
        (r"$L_{focal}$", "Focal Loss: Focuses on hard-to-classify examples (class imbalance)"),
        (r"$L_{fbeta}$", "F-Beta Loss: Prioritizes Recall over Precision (catch more avalanches)"),
        (r"$L_{recall}$", "Recall Loss: Specifically penalizes False Negatives (missed dangers)"),
        (r"$L_{physics}$", "Physics Loss: Penalty for violating energy balance equation")
    ]
    
    # Colors for the bullets/math text
    colors = ['#b71c1c', 'black', '#ef6c00', '#1565c0', '#2e7d32', '#6a1b9a', '#33691e']

    for i, (symbol, desc) in enumerate(variables):
        y_pos = y_start - i * y_step
        
        # Draw symbol with colored background bubble
        circ = patches.FancyBboxPatch((0.5, y_pos - 0.3), 1.2, 0.6, boxstyle="round,pad=0.1", 
                                      linewidth=0, facecolor='white', alpha=0.8)
        ax.add_patch(circ)
        
        ax.text(1.1, y_pos, symbol, fontsize=16, fontweight='bold', ha='center', va='center', color=colors[i])
        ax.text(2.0, y_pos, f":  {desc}", fontsize=14, ha='left', va='center', color='#0d1b2a')

    plt.tight_layout()
    plt.savefig('avalanche_loss_function_breakdown.png', dpi=300, bbox_inches='tight', transparent=True)
    plt.show()

draw_loss_function_breakdown()
```
